```python
import os
import csv
from music21 import converter, chord

# Define the path to the main folder containing the MIDI files
main_folder_path = 'F:/datasetn'

# Define the path for the output CSV file
csv_file_path = 'F:/datasetn/test1.csv'

# Open a new CSV file to write the data
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    # Create a CSV writer object
    csv_writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    # Write the header row
    csv_writer.writerow(['Chord', 'Notes'])
    
    # Iterate over all files in the main folder
    for subdir, dirs, files in os.walk(main_folder_path):
        for file in files:
            # Check if the file is a MIDI file
            if file.endswith('.mid') or file.endswith('.midi'):
                # Construct the full path to the file
                file_path = os.path.join(subdir, file)
                print(f"Processing file: {file_path}")
                
                # Load the MIDI file
                midi_file = converter.parse(file_path)
                
                # Iterate through all elements and find chords
                for element in midi_file.recurse():
                    if isinstance(element, chord.Chord):
                        # Get the chord's name (e.g., D minor)
                        chord_name = element.pitchedCommonName
                        
                        # Get the individual notes of the chord without octaves
                        note_names = {p.name for p in element.pitches}  # Use a set to store unique note names
                        
                        # Write the chord and its notes to the CSV file
                        csv_writer.writerow([chord_name, ' '.join(sorted(note_names))])  # Sort the note names for consistent ordering
                        
print(f"Data Saved")
```

```python
import os
import glob
import mido
import pandas as pd
from mido import MidiFile

# Function to extract notes from a MIDI track
def extract_notes(midifile):
    mid = MidiFile(midifile)
    notes = []

    for i, track in enumerate(mid.tracks):
        for msg in track:
            if not msg.is_meta and msg.type == 'note_on' and msg.velocity > 0:
                note = msg.note  # MIDI note number
                notes.append(note)
    return notes

# Function to save the extracted notes to a CSV file
def save_to_csv(data, filename):
    df = pd.DataFrame(data, columns=['Filename', 'Notes'])
    df.to_csv(filename, index=False)
    print(f"Data saved to {filename}")

# Directory containing MIDI files and subdirectories
midi_dir = 'F:/Course/thesis topic/resources materials/datasets/Midi files'  # Replace with the path to your MIDI directory
output_csv = 'output_notes.csv'  # The CSV file where the data will be saved

# List to hold data for all MIDI files
all_midi_data = []

# Traverse directory and process all MIDI files
for root, dirs, files in os.walk(midi_dir):
    for file in files:
        if file.endswith('.mid') or file.endswith('.midi'):
            midi_path = os.path.join(root, file)
            notes = extract_notes(midi_path)
            all_midi_data.append({'Filename': midi_path, 'Notes': notes})

# Save the extracted data to a CSV file
save_to_csv(all_midi_data, output_csv)
```
