```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("mudassar93/llama2-chat-piano")
model = AutoModelForCausalLM.from_pretrained("mudassar93/llama2-chat-piano")
```

```python
from IPython.display import display, Markdown
```

```python
def stream(user_prompt):
    #runtimeFlag = "cuda:0"
    system_prompt = 'You are a helpful assistant that provides accurate and concise responses realted to music theory'
    #system_prompt = 'You are an helpful assistant trained in music theory. Your task is to provide accurate, concise, and informative responses  to user query.'


    B_INST, E_INST = "[INST]", "[/INST]"
    B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"

    prompt = f"{B_INST} {B_SYS}{system_prompt.strip()}{E_SYS}{user_prompt.strip()} {E_INST}\n\n"

    inputs = tokenizer([prompt], return_tensors="pt")

    streamer = TextStreamer(tokenizer)

    # Despite returning the usual output, the streamer will also print the generated text to stdout.
    _ = model.generate(**inputs, streamer=streamer, temperature= 0.2, max_new_tokens=348)
```

```python
from music21 import stream, note, midi

# Define the melody
melody = "C G C E G C E G C E G C".split()

# Create a music21 stream
s = stream.Stream()

# Add notes to the stream
for pitch in melody:
    new_note = note.Note(pitch)
    new_note.duration.quarterLength = 1  # Quarter note
    s.append(new_note)

# Save to MIDI file
midi_file_path = 'melody1.mid'
mf = midi.translate.streamToMidiFile(s)
mf.open(midi_file_path, 'wb')
mf.write()
mf.close()
```

```python
from midi2audio import FluidSynth
from pydub import AudioSegment

# Convert MIDI to WAV using FluidSynth
fs = FluidSynth()
audio_file_path = 'melody1.wav'
fs.midi_to_audio(midi_file_path, audio_file_path)

# Convert WAV to MP3 (optional, requires ffmpeg)
!apt install ffmpeg
audio_mp3_path = 'melody1.mp3'
AudioSegment.from_wav(audio_file_path).export(audio_mp3_path, format="mp3")
```

```python
from langchain.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.prompts.prompt import PromptTemplate

# Template using jinja2 syntax
template = """
<s>[INST] <<SYS>>
You are a helpful assistant trained in music theory, you always only answer for the assistant then you stop. Read the chat history to get context.
<</SYS>>

Current conversation:
{{ history }}

{% if history %}
    <s>[INST] Human: {{ input }} [/INST] AI: </s>
{% else %}
    Human: {{ input }} [/INST] AI: </s>
{% endif %}
"""

prompt = PromptTemplate(
    input_variables=["history", "input"],
    template=template,
    template_format="jinja2"
)
```

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory

# Initialize the conversation chain
conversation = ConversationChain(
    llm=llm,
    memory=ConversationBufferMemory(),
    prompt=prompt,
    verbose=False
)
```

```python
# Start the conversation
def predict(message: str, history: str):
    response = conversation.predict(input=message)

    return response
```

```python
import gradio as gr

# Set up the user interface
interface = gr.ChatInterface(
    clear_btn=None,
    fn=predict,
    retry_btn=None,
    undo_btn=None,
)

# Launch the user interface
interface.launch(
    height=600,
    inline=True,
    share=True,
    width=800
)
```
